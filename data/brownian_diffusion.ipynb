```python
import numpy as np
import matplotlib.pyplot as plt
import imageio

N = 100
T = 500

points = np.zeros((N, 2))
frames = []

fig, ax = plt.subplots(figsize=(6, 6))

for t in range(T):
    # Brownian motion step
    points += np.random.randn(N, 2)*0.3 # multiplying by 0.3 as if we dont do this the points spread out way to fast

    if t % 10 == 0:
        ax.clear()
        ax.scatter(points[:, 0], points[:, 1], s=5)
        ax.set_title(f"Brownian motion (t={t})")
        ax.set_xlim(-10, 10)
        ax.set_ylim(-10, 10)

        fig.canvas.draw()

    
        image = np.asarray(fig.canvas.buffer_rgba()).copy()
        frames.append(image)

plt.close(fig)

imageio.mimsave("brownian_motion.gif", frames, fps=5 , loop=0)

```
